```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Import dataset

```python
dataset = pd.read_csv('Position_Salaries.csv')
```

```python
x = dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values
```

```python
x
```

Training the Linear Regression model on the whole dataset

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x,y)
```

Training the Polynominal Regression model on the whole dataset

```python
from sklearn.preprocessing import PolynomialFeatures
# This module from sklearn.preprocessing is used to transform the original feature x into
# a higher-dimensional space by adding polynomial terms.
from sklearn.linear_model import LinearRegression

poly_reg = PolynomialFeatures(degree=2)
# The degree=2 means it will generate features up to the second degree
#For example, if x = [x1, x2, x3], the transformed features x_poly will include [1, x1, x2, x1^2, x1*x2, x2^2].

# Transform the input feature x into polynomial features:
x_poly = poly_reg.fit_transform(x)
# After this step, x_poly will contain the original data plus the new polynomial features.
model_poly = LinearRegression()
model_poly.fit(x_poly,y)

# This code is essentially performing polynomial regression, which can be used when the relationship between
# x and y is not linear but can be modeled using polynomial terms (like quadratic or cubic). Even though LinearRegression
# is being used, it's applied to the transformed polynomial features, enabling the model to fit non-linear patterns.
```

Visualizing the Linear Regression results

```python
plt.scatter(x,y,color='red')
plt.plot(x,model.predict(x),color='blue')
plt.title('Salary based on Position level_ Linear model')
plt.xlabel('position level')
plt.ylabel('Salary')
plt.show()
```

Visualizing the Polynominal Regression results

```python
plt.scatter(x,y,color='red')
plt.plot(x,model_poly.predict(x_poly),color='blue')
plt.title('Salary based on Position level_ Polynominal model')
plt.xlabel('position level')
plt.ylabel('Salary')
plt.show()
```

Visualizing the Polynominal Regression results(for higher resolution and smoother curve)

```python
#Lets build a new model and increase the degree:
poly_reg2 = PolynomialFeatures(degree=4)
x_poly2 = poly_reg2.fit_transform(x)

model_poly2 = LinearRegression()
model_poly2.fit(x_poly2,y)
```

```python
plt.scatter(x,y,color='red')
plt.plot(x,model_poly2.predict(x_poly2),color='blue')
plt.title('Salary based on Position level_ Polynominal model2')
plt.xlabel('position level')
plt.ylabel('Salary')
plt.show()
```

Predicting a new result with Linear Regression

```python
model.predict([[6.5]])
```

Predicting a new result with Polynominal Regression

```python
model_poly2.predict(poly_reg2.fit_transform([[6.5]]))
```
